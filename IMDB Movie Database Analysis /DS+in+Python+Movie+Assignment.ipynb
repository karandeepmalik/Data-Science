-  ### Subtask 2.3: Drop unecessary rows using columns with high Null percentages

Now, on inspection you might notice that some columns have large percentage (greater than 5%) of Null values. Drop all the rows which have Null values for such columns.

```python
# Code for dropping the rows here
# On inspection we found that gross and budget columns had large no of null values
# So, we are slicing of the null vlaues for both these columns isnan check if the value is NaN and
# ~ ensures the reverse of the conition i.e. columns with non-NaN values are discarded from data frame
movies =  movies[~np.isnan(movies['gross'])]
movies =  movies[~np.isnan(movies['budget'])]
print(round(100*(movies.isnull().sum()/len(movies.index)), 2))
```

-  ### Subtask 2.4: Drop unecessary rows

Some of the rows might have greater than five NaN values. Such rows aren't of much use for the analysis and hence, should be removed.

```python
# Code for dropping rows that have greater than five NaN values
# isnull() returns the rows with null values for column and sum() funciton gets us the count of such rows
# The condition less than 5 ensures we select only those rows with <5 nulls for columns
movies = movies[movies.isnull().sum(axis=1) <= 5]
print(round(100*(movies.isnull().sum()/len(movies.index)), 2))
```

## Task 3: Data Analysis

-  ### Subtask 3.1: Change the unit of columns

Convert the unit of the `budget` and `gross` columns from `$` to `million $`.

```python
# The current unit of budget and gross is $. Dividing by 10,00,000 will gives us values in millions
# The nanonymous lambda function divides each value by 1 million and converts the values into million dollars
movies['budget'] = movies['budget'].apply(lambda x: x/1000000)
movies['gross'] = movies['gross'].apply(lambda x: x/1000000)
```

-  ### Subtask 3.2: Find the movies with highest profit

    1. Create a new column called `profit` which contains the difference of the two columns: `gross` and `budget`.
    2. Sort the dataframe using the `profit` column as reference.
    3. Extract the top ten profiting movies in descending order and store them in a new dataframe - `top10`

```python
# Code for creating the profit column here
# Profit = Gross Earnings - Budget for making movies
# The below code will create a new column 'profit' and calculate its value by subtracting the value in gross column
# with value in budget column
movies['profit'] = movies['gross'] - movies['budget']
print(movies.head(10))
```

```python
# Code for sorting the dataframe
# The sort_values() function when used with a data frame sorts the data frame rows according to the column
# specified with 'by' parameter. ascending = False means it will osrt in descending order. 
movies = movies.sort_values(by=['profit'],ascending=False)
print(movies.head())
```

```python
# Code to get the top 10 profiting movies 
# Since in the previous steps we have already introduced a column profit, assigned it appropriate values
# and sorted data frame according to profit column, the top 10 values for 'movie_title' will gives us the top
# 10 profiting movies. To get the top 10 values, we use the head() function with parameter '10'.
top10 = movies.head(10)['movie_title']
print(top10)
```

    1. Create a new dataframe `IMDb_Top_250` and store the top 250 movies with the highest IMDb Rating (corresponding to the column: `imdb_score`). Also make sure that for all of these movies, the `num_voted_users` is greater than 25,000.

-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Code for extracting the first two genres of each movie here
# The apply() function is used to apply a function repeatedly over a given series.

# Inorder to find the genre_1, for each row  we will split the existing genres column value
# using separator '|' and take the first substring obtained
movies['genre_1'] = movies['genres'].apply(lambda x: x.split('|')[0] )

# Inorder to find the genre_2, for each row  we will split the existing genres column value
# using separator '|' and take the second substring obtained if available or use the first substring
# if not available and insert it into 'genre_2' column
movies['genre_2'] = movies['genres'].apply(lambda x: x.split('|')[1] if len(x.split('|')) > 1 else x.split('|')[0] )
print(movies.head(50))

# As you can see, for movie Hangover which had only one 'genres' value 'Comedy', both 'genre_1' and 'genre_2'
# are comedy.
```

```python
# We will create a new variable movies_by_segmentto represent the grouping of movies dataframe by 'genre_1'
# and genre_2
movies_by_segment = movies.groupby(['genre_1', 'genre_2'])
```

```python
# As per the subtask requirement, we will now obtain the Top 5 most popular genre_1 + genre_2 Combo 
# For this we will find the mean of gross values for each 'genre_1' and 'genre_2' combo, sort the values
# and the the top 5 Combo using head() function
PopGenre =  movies_by_segment['gross'].mean().sort_values(ascending=False).head(5)
print(PopGenre)

# As you can see below Family + Sci-Fi is the most popular combo of genres.
```

**Checkpoint 5:** Well, as it turns out. `Family + Sci-Fi` is the most popular combo of genres out there!

-  ### Subtask 3.7: Find the critic-favorite and audience-favorite actors

    1. Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_user_for_review` and identify the actors which have the highest mean.

```python
# Code for creating Meryl_Streep data frame
# We need to find all movies in which 'Meryl Streep' was the lead. The lead actor is the value represented 
# by actor_1_name column
Meryl_Streep =  movies.loc[movies['actor_1_name'] == 'Meryl Streep']
print(Meryl_Streep.head(10))
```

```python
# Code for creating Leo_Caprio data frame
# We need to find all movies in which 'Leonardo DiCaprio' was the lead. The lead actor is the value represented 
# by 'actor_1_name' column

Leo_Caprio = movies.loc[movies['actor_1_name'] == 'Leonardo DiCaprio']
print(Leo_Caprio.head(10))
```

```python
# Code for creating Leo_Caprio data frame
# We need to find all movies in which 'Brad Pitt' was the lead. The lead actor is the value represented 
# by 'actor_1_name' column
Brad_Pitt = movies.loc[movies['actor_1_name'] == 'Brad Pitt'] 
print(Brad_Pitt.head(10))
```

```python
# Code for combining the three dataframes here
# We will now combined the obtained above three data frames to get a new data frame 'Combined'
# We will use the append function to append the values of each actor one after another.
Combined = (Meryl_Streep.append(Leo_Caprio)).append(Brad_Pitt )
print(Combined.head(20))
```

```python
# Code for grouping the combined dataframe by 'actor_1_name'
# We will use groupby function to obtain this and assign it to variable 'groupby_1_name' 
groupby_1_name = Combined.groupby('actor_1_name')
```

```python
# Code for finding the mean of critic reviews and audience reviews here

# Now using the above groupby result we will find the mean of num of critic reviews and assign it to 
#'mean_num_critic_reviews'
mean_num_critic_reviews = groupby_1_name.num_critic_for_reviews.mean() 

# Similarly, using the groupby result we will find the mean of num of user reviews and assign it to 
#'mean_num_user_reviews'
mean_num_user_reviews = groupby_1_name.num_user_for_reviews.mean() 

# Now, after sorting the values based on critical reviews we will try to obtain the actor name with 
# highest mean value of critic reviews
actor_highest_num_critic_reviews = mean_num_critic_reviews.sort_values(ascending=False).index[0]

# Now, after sorting the values based on user reviews we will try to obtain the actor name with 
# highest mean value of user reviews
actor_highest_num_user_reviews = mean_num_user_reviews.sort_values(ascending=False).index[0]

print(actor_highest_num_critic_reviews)
print(actor_highest_num_user_reviews)

# As you can see below Leonardo Di Caprio tops both the lists.
```

**Checkpoint 6:** `Leonardo` has aced both the lists!
